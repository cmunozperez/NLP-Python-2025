Otra cosa que podemos averiguar es quiénes son los autores con la mayor cantidad de manuscritos en el repositorio. Para ello, vamos a necesitar una función que opere sobre el DataFrame y devuelva un diccionario que nos diga el número de manuscritos de cada autor.

```python
from collections import Counter

# A function to get the number of manuscripts per author
def ms_per_author(n = None):
    '''
    Takes a number n and returns the n authors with the most manuscripts.
    If no n is introduced, it returns a dictionary with counts for all authors.

    Parameters
    ----------
    n : int, optional

    Returns
    -------
    dictionary
        A dictionary with counts of manuscripts per author

    '''
    authors = []
    for cell in df_main['Authors']:
        authors += cell.split(', ')
    if n == None:
        return dict(Counter(authors))
    else:
        return dict(Counter(authors).most_common(n))
```

A continuación, se usa la función ``ms_per_author()`` para encontrar los diez autores con más artículos.

```python
ten_most_contributions = ms_per_author(10)

ten_most_contributions
```

También podemos calcular las descargas totales de cada autor. Primero armé una función para calcular esto a partir de los top *n* autores más descargados.

```python
def downloads_per_author(n=None):
    '''
    Returns the n authors with the most downloads from df_main.

    Parameters
    ----------
    n : int, optional
        Number of top authors to return. If None, returns all.

    Returns
    -------
    dict
        Dictionary with download counts per author.
    '''
    authors_down = {}

    for i in range(len(df_main)):
        authors_raw = df_main.iloc[i, 2]
        downloads = df_main.iloc[i, 6]

        if pd.isna(authors_raw) or pd.isna(downloads):
            continue

        aut_paper = [a.strip() for a in authors_raw.split(', ')]

        for person in aut_paper:
            authors_down[person] = authors_down.get(person, 0) + int(downloads)

    if n is None:
        n = len(authors_down)

    sorted_n = sorted(authors_down.items(), key=lambda x: x[1], reverse=True)[:n]
    return {k: int(v) for k, v in sorted_n}
```

Acá usamos la función `downloads_per_author()` para obtener los nombres de los diez autores con la mayor cantidad de descargas.

```python
ten_most_downloads = downloads_per_author(10)

ten_most_downloads
```

## Palabras clave y wordcloud

La mayoría de los manuscritos de la base de datos tienen palabras clave que describen su contenido. Es posible contar las palabras clave más frecuentes en el repositorio para ver qué tópicos son los más "populares" (es lingüística; la respuesta estricta es "ninguno"). En primer lugar, creamos una función que cuente el uso de cada palabra clave.

```python
def keyword_counter(n=None):
    '''
    Returns the n most common keywords from df_main['Keywords'].
    If n is not given, returns all counts.

    Parameters
    ----------
    n : int, optional

    Returns
    -------
    dict
        Dictionary of keyword counts.
    '''
    keywords = []

    for cell in df_main['Keywords']:
        if pd.isna(cell):
            continue
        for kw in cell.split(','):
            kw_clean = kw.strip().lower()
            if kw_clean: # this filters empty elements
                keywords.append(kw_clean)

    counts = Counter(keywords)

    if n is None:
        return dict(counts)
    else:
        return dict(counts.most_common(n))
```

Acá uso la función `keyword_counter()` para obtener las diez palabras clave más frecuentes. Los resultados reflejan que, para cargar un manuscrito, es necesario definir a qué subdisciplina lingüística pertenece, es decir, sintaxis, semántica, morfología o fonología.

```python
top_10_keywords = keyword_counter(10)

top_10_keywords
```

Una mejor manera de visualizar la frecuencia de las palabras clave es mediante una nube de palabras. El siguiente código genera una nube con las 150 palabras clave más frecuentes. ## Tendencias de las subdisciplinas a lo largo del tiempo

Dado que la mayoría de los artículos incluyen palabras clave relacionadas con sus respectivas subdisciplinas lingüísticas (es decir, sintaxis, semántica, morfología y fonología), podemos cuantificar el volumen de artículos asociados a cada subdisciplina a lo largo de los años. Esto permite observar la evolución de la popularidad de cada subdisciplina.

El primer paso consiste en crear una función que seleccione subdisciplinas lingüísticas a partir de una lista de palabras clave.

```python
# This is a function that applied to a cell with keywords returns the linguistic subdiscipline listed there (e.g., syntax)
# or a combination of them if two or more subdisciplines are listed (e.g., morphology + syntax)

def keyword_checker(cell):
    """
    Given a string of comma-separated keywords, returns a cleaned discipline label.
    If the input is missing or doesn't match any target keyword, returns ''.
    """
    if pd.isna(cell):
        return ''

    target_keywords = ['syntax', 'semantics', 'phonology', 'morphology']
    contained_keywords = []

    for word in target_keywords:
        if word.lower() in cell.lower():
            contained_keywords.append(word)

    replacement = ' + '.join(contained_keywords)
    return replacement
```

Esta función permite crear un nuevo DataFrame con la cantidad de papers con cada combinación de subdisciplinas para cada año.

```python
# Create a new DataFrame for analysis
df_key = pd.DataFrame()
df_key['Id'] = df_main['Id']

# Extract year from the date column
df_key['Year'] = df_main['Date'].dt.year.astype('Int64').astype(str)

# Apply keyword_checker to extract main disciplines
df_key['Disciplines'] = df_main['Keywords'].apply(keyword_checker)

# Group by year and discipline, and count manuscripts
df_key = df_key.groupby(['Year', 'Disciplines']).size().reset_index(name='count')

df_key
```

Esta tabla no sirve de mucho, pero la vamos a usar para crear una tabla un poco más informativa. Vamos paso por paso:

- Primero, saquemos las combinaciones de disciplinas menos atestiguadas.
- Vamos a reorganizar estos datos en una tabla que tenga AÑOS como filas y DISCIPLINAS como columnas.
- Vamos a restringir los años al período 2002-2022
- Reordenamos las columnas (no importa, realmente)

```python
# Drop rare or noisy combinations (quite ad hoc, to be honest)
combinations_to_drop = [
    '',
    'morphology + semantics',
    'phonology + morphology + semantics',
    'phonology + morphology + syntax',
    'phonology + morphology + syntax + semantics',
    'phonology + semantics',
    'phonology + syntax',
    'phonology + syntax + semantics'
]

df_key = df_key[~df_key['Disciplines'].isin(combinations_to_drop)]

# Pivot table: years as rows, disciplines as columns
df_key = df_key.pivot_table(index='Year', columns='Disciplines', values='count', fill_value=0)

# --- Drop years before 2002 and remove incomplete year (e.g. 2023) ---
years_to_drop = [str(y) for y in range(0, 2002)] + ['2023']
df_key = df_key[~df_key.index.isin(years_to_drop)]

# --- Reorder columns (only if they exist) ---
desired_order = [
    'syntax',
    'semantics',
    'syntax + semantics',
    'morphology + syntax',
    'phonology',
    'morphology + syntax + semantics',
    'morphology',
    'phonology + morphology'
]

existing_order = [col for col in desired_order if col in df_key.columns]
df_key = df_key[existing_order]

# --- Show the final result ---
df_key.head(21)
```

Esta tabla sirve para hacer un gráfico.

```python
# This plots a graph from df_key
df_key.plot(kind='line', stacked=False, figsize=(16, 9), linewidth=3)
plt.xlabel('')
plt.title('Keyword trends in Lingbuzz (2002-2022)', fontsize=18)
plt.legend(fontsize=12);
```

## Colaboraciones y redes de coautoría

La información sobre la autoría de los manuscritos puede utilizarse para identificar las colaboraciones más fructíferas dentro del repositorio, así como para construir redes de coautoría dentro del campo de la lingüística.

La siguiente función recupera pares de autores en el conjunto de datos y el número de manuscritos que escribieron juntos.

```python
# This function returns a dictionary with a pair of authors (tuple) as key, and the number of papers they wrote together as value
def network_connections():
    connections = {}
    for cell in df_main['Authors']:
        authors = cell.split(', ')
        conn_by_paper = []
        for a in authors:
            for b in authors:
                if a != b:
                    conn = frozenset({a, b})
                    conn_by_paper.append(conn)
        conn_by_paper = set(conn_by_paper)
        for i in conn_by_paper:
            if i in connections.keys():
                connections[i] += 1
            else:
                connections[i] = 1
    return {tuple(key): value for key, value in connections.items()}
```

```python
connections = network_connections()
```

The following retrieves the pairs of authors who wrote together 9 or more manuscripts in Lingbuzz.

```python
# This filters the pairs of authors who wrote together more than 8 papers
for people, ms_together in connections.items():
    if ms_together > 8:
        print(people, ms_together)
```

Esta información puede utilizarse para construir una *red de coautoría*. Considero que son grafos no dirigidos, como se indica en la definición de [Savić, Ivanović, & Jain (2019: 181)](https://doi.org/10.1007/978-3-319-91196-0_5).

> Let *U* be a set of bibliographic units (journal and conference publications, books, and so on), and let *A* be the set of authors appearing in *U*. The co-authorship network corresponding to *U* is most commonly defined as an undirected and weighted graph *G = (V, E)* with the following properties:

> - The set of nodes *V* corresponds to the set of authors *A*, i.e. each author from *A* is represented by one node in *G*.
> - Two authors *x* and *y* are connected by an undirected link *e* (*e ∈ E*) if there is at least one bibliographic unit in *U* jointly co-authored by *x* and *y* (with or without other co-authors). In other words, co-authorship networks are author-centered onemode projections of bipartite networks linking researchers to bibliographic units they (co-)authored.
> - Link weights express the strength of collaboration between connected authors.

```python
import networkx as nx

# This creates an undirected graph
G = nx.Graph()

# This "fills it" with the data from the pairs of authors
for key, value in connections.items():
    G.add_edge(key[0], key[1], weight=value)
```

El siguiente código devuelve el número total de autores en lingbuzz.



```python
total_amount_authors = G.number_of_nodes()

total_amount_authors
```

Los siguientes son los autores con más co-autorías en el repositorio.

```python
#Getting the people with most co-authors
most_collaborators = sorted(list(G.degree()), key=lambda x: x[1], reverse=True)
most_collaborators[:10]
```

El siguiente paso consiste en comprobar si las relaciones de coautoría determinan una única red o si existen varias redes de coautoría en el repositorio.

```python
only_one_network = nx.is_connected(G)

print('Is there a unified co-authorship network among the authors in Lingbuzz?')
print(only_one_network)
```

El siguiente código recupera (i) el número total de redes de coautoría en lingbuzz, y (ii) el número de autores en cada una de las cinco redes más grandes.

```python
co_authorship_networks = list(nx.connected_components(G))

size_networks = [len(network) for network in co_authorship_networks]

print(f'There are {len(size_networks)} co-authorship networks in Lingbuzz.')
print(f'The five biggest ones have the following number of authors in them: {sorted(size_networks, reverse=True)[:5]}')
```

Así, 1530 de un total de 2225 autores forman parte de la misma red de coautoría. El podemos representar esta red con un grafo. La imagen producida aquí es más pequeña para ahorrar tiempo. Una versión más grande y legible de este gráfico puede descargarse [aquí](https://github.com/cmunozperez/lingbuzz_data_analysis/blob/main/images/output_44_1.png).

```python
# This takes the largest co-authorship network
biggest_network = max(co_authorship_networks, key=len)

# This creates a subgraph with it
subgraph = G.subgraph(biggest_network)

# And this generates a plot form it


## This produces a smaller image
plt.figure(figsize=(36, 24))

pos = nx.spring_layout(G, k=0.2)
nx.draw(subgraph, pos, with_labels=True, node_color='skyblue', node_size=50, font_size=3,
        font_color='red', font_weight='bold', edge_color='darkgrey')



# The following code generates an image big enough to have readable labels, but slow to build
###########################################################################
# plt.figure(figsize=(216, 144))

# pos = nx.spring_layout(G, k=0.2)
# nx.draw(subgraph, pos, with_labels=True, node_color='skyblue', node_size=200, font_size=15,
#         font_color='red', font_weight='bold', edge_color='darkgray')
###########################################################################


plt.show();
```
